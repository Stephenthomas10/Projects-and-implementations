```python
#1.Scaling transformation: multiplying a vector by a scalar
import numpy as np

# define the scaling factor
s = 2

# define the vector to be scaled
v = np.array([1, 2, 3])

# perform the scaling transformation
v_scaled = s * v

# print the original and scaled vectors
print("Original vector:", v)
print("Scaled vector:", v_scaled)
```

```python
#affine
import numpy as np
import matplotlib.pyplot as plt

# Define the scaling factor and translation vector
s = 2
t = np.array([1, 2])

# Define the matrix A
A = np.array([[2, 1], [1, 2]])

# Define the original rectangle vertices
vertices = np.array([[0, 0], [0, 1], [1, 1], [1, 0]])

# Add a column of ones to the vertices array to enable matrix multiplication with the affine transformation matrix
vertices_homogeneous = np.hstack((vertices, np.ones((4, 1))))

# Perform the affine transformation
vertices_transformed_homogeneous = np.dot(vertices_homogeneous.T, A.T) + t

# Remove the homogeneous coordinate from the transformed vertices
vertices_transformed = vertices_transformed_homogeneous[:, :2]

# Plot the original and transformed rectangles
fig, ax = plt.subplots()
ax.add_patch(plt.Polygon(vertices, fill=None, edgecolor='b', linewidth=2, label='Original'))
ax.add_patch(plt.Polygon(vertices_transformed, fill=None, edgecolor='r', linewidth=2, label='Transformed'))
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Affine Transformation')
ax.legend()

plt.show()


```

```python
#translation
import numpy as np
import matplotlib.pyplot as plt

# Translation matrix
T = np.array([[1, 0, 2], [0, 1, 1], [0, 0, 1]])

# Original points
X = np.array([[1, 2, 2, 1], [1, 1, 2, 2], [1, 1, 1, 1]])

# Transformed points
X_transformed = T @ X

# Plot original and transformed points
plt.figure(figsize=(5,5))
plt.plot(X[0], X[1], 'b-', label='Original')
plt.plot(X_transformed[0], X_transformed[1], 'r--', label='Transformed')
plt.legend()
plt.show()
```

```python
#rotation
import numpy as np
import matplotlib.pyplot as plt

# Rotation matrix
theta = np.pi/4
R = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])

# Original points
X = np.array([[1, 2, 2, 1], [1, 1, 2, 2]])

# Transformed points
X_transformed = R @ X

# Plot original and transformed points
plt.figure(figsize=(5,5))
plt.plot(X[0], X[1], 'b-', label='Original')
plt.plot(X_transformed[0], X_transformed[1], 'r--', label='Transformed')
plt.legend()
plt.show()
```

```python
#shearing
import numpy as np
import matplotlib.pyplot as plt

# Shearing matrix
Sh = np.array([[1, 1], [0, 1]])

# Original points
X = np.array([[1, 2, 2, 1], [1, 1, 2, 2]])

# Transformed points
X_transformed = Sh @ X

# Plot original and transformed points
plt.figure(figsize=(5,5))
plt.plot(X[0], X[1], 'b-', label='Original')
plt.plot(X_transformed[0], X_transformed[1], 'r--', label='Transformed')
plt.legend()
plt.show()
```
