```python
class Character:
  # Task 1: Create a class Character with the following attributes
  # *   name
  # *   maxHP
  # *   HP (This is equal to MaxHP)
  # *   attack
  # *   defense
  # *   maxMP
  # *   MP (This is equal to maxMP)
  # *   inventory = [] <- This is not a parameter. This is an empty list by default.

  def __init__(self, name, maxHP, attack, defense, maxMP):
    #-- Your code here --#
    self.name = name
    self.maxHP = maxHP
    self.HP = maxHP
    self.attack = attack
    self.defense = defense
    self.maxMP = maxMP
    self.MP = maxMP
    self.inventory = []

  def __str__(self):
    return f"""{self.name}
    HP:{self.HP}/{self.maxHP}
    MP:{self.MP}/{self.maxMP}"""

  # Task 2: Create a function receiveDamage, that takes an 'damage' parameter.
  # The function signature has been given to you. You need to add functionality.
  # Within the function you must perform the following calculation

  # Reduce self.HP by the self.damage.
  # Have a check to see if self.HP < 0. In this case,
  # Print a message saying '{self.name} lost {receivedDamage} HP!'.
  # Finally return the receivedDamage

  def receiveDamage(self, damage):
    # Step 1: Create a new variable receivedDamage and assign its value as damage minus the character's defense.
    # Your code here
    receivedDamage = damage - self.defense

    # Step 1.5: If your receivedDamage < 0, then change it to 0.
    if receivedDamage < 0:
      receivedDamage = 0

    # Step 2: Subtract HP by damage received and check if your current HP is below 0. If it is below 0, set HP to 0.
    # Your code here
    self.HP -= receivedDamage
    if self.HP < 0:
      self.HP = 0

    # Do not change the code below. It just prints the results.
    print(f"{self.name} lost {receivedDamage} HP! Current HP = {self.HP}")
    return receivedDamage

  # Task 3: Create a setter method(function) getAttack that just returns the attack of the character.

  def getAttack(self):
    # Your code here
    return self.attack

  # Task 4: Create a method(function) called isAlive(). This
  # will take no parameters(besides self) and return a boolean value True if
  # HP > 0 and False otherwise.

  def isAlive(self):
    # Your code here
    if self.HP > 0:
      return True
    return False

  # Task 5: Create a method called displayInventory.
  # Step 1: This function will print the
  #         index and value of every item in the inventory list.
  #         Hint: Use a for in range to retrieve each index and print the corresponding value.
  #         A sample output will look like the below, supposing inventory = ['Potion', 'Elixir']
  #         0 : Potion
  #         1 : Elixir
  # Step 2: return the length of the inventory list

  def displayInventory(self):
    # Your code here
    for index in range(0, len(self.inventory)):
      print(index, self.inventory[index])
    return len(self.inventory)

  # DO NOT modify the healHP function
  def healHP(self, effect):
    self.HP += effect
    if self.HP > self.maxHP:
      self.HP = self.maxHP

  # Using the healHP function as an example, modify the function healMP that will
  # increase the Character's current MP by effect.
  # If the character's current MP is greater than maxMP, set the value to maxMP.
  def healMP(self, effect):
    # Your code here
    self.MP += effect
    if self.MP > self.maxMP:
      self.MP = self.maxMP

  # You do not have to modify any code below. This has already been implemented.
  # You will be using this code on a later basis, so just try to understand it.
  def consumeItem(self, index):
    # This function will consume the item at the mentioned index.
    if index in range(0, len(self.inventory)):
      item = self.inventory[index]
      if item.name == "Potion":
        self.healHP(item.effect)
        print(f"Potion consumed. {item.effect} HP healed")
      elif item.name == "Elixir":
        self.healMP(item.effect)
        print(f"Elixir consumed. {item.effect} MP healed")
      self.inventory.remove(item)

  def addItem(self, item):
    # This function will add a new item into the inventory, if item is not None
    if item == None:
      return
    elif type(item) == list:
      self.inventory += item
    else:
      self.inventory.append(item)
```

```python
# Task 5: Finish the following code to make Class Warrior inherit the Character Class.
class Warrior(Character):
  # Now, the Warrior class will have the following attributes that are fixed.
  # * MaxHP = HP = 100
  # * Attack = 25
  # * Defense = 10
  # * MaxMP = MP = 5

  def __init__(self, name):
    # Initialize the values using the super constructor.
    # Your code here
    super().__init__(name, 100, 25, 10, 5)

  def slash(self):
    # This slash represents a special attack. And it will return an attack value
    # that is twice the character's current attack value.
    if self.MP > 2:
      print("SLASH!!!!!")
      self.MP -= 2
      return self.attack * 2
    else:
      print("You do not have enough MP")
      print("Current MP :", self.MP)
      print("Instead, using a normal attack")
      return self.attack
```

```python
# The Mage class will have the following attributes that are fixed.
# * maxHP = HP = 80
# * attack = 10
# * defense = 5
# * maxMP = 100

# It will have the following methods
# Step 1: __init__(self, name) <- The constructor
# Step 2: fireball(self) <- The special attack. This is very similar to slash().
#         This function will return the character's attack * 6 if it has greater than 10 MP. Also, deduct 10 MP from your current MP.
#         Else, your class will have to use a normal attack. Also print a message for the same.
# Hint: This class is almost exactly a ditto copy to the Warrior Class. Feel free to copy, paste and edit as required.

#-- Your code here --#
# Task 5: Finish the following code to make Class Warrior inherit the Character Class.
class Mage(Character):
  def __init__(self, name):
    # Initialize the values using the super constructor.
    # Your code here
    super().__init__(name, 80, 10, 5, 100)

  def fireball(self):
    if self.MP > 10:
      print("FIREBOLT!!!!!")
      self.MP -= 10
      return self.attack * 6
    else:
      print("You do not have enough MP")
      print("Current MP :", self.MP)
      print("Instead, using a normal attack")
      return self.attack
```

```python
# Now, you shall create a class Enemy, which inherits Character.

# Also, create a function enraged, which updates its attack using the following equation only if the enemy's HP < 10.
# enemy's attack = enemy's attack * 3
class Enemy(Character):

  # It's constructor should accept all the parameters and call it's parent constructor using the super() keyword.
  # You have also been given an extra carrying parameter. Now this value should be set to self.carrying
  def __init__(self, name, maxHP, attack, defense, maxMP, carrying):
    # Your code here
    super().__init__(name, maxHP, attack, defense, maxMP)
    self.carrying = carrying

  # You do not need to modify this
  def __str__(self):
    return f"""Enemy: {self.name}.
    HP: {self.HP}/{self.maxHP}"""

  # This is a setter method(function). When the enemy is in an enraged status, this function will be called.
  # This function will multiple the enemy's attack by 3 and set that as the new attack.
  def enraged(self):
    # Your code here
    self.attack *= 3

  # Finally, we need to have a check to see if the enemy is enraged. The logic work like the following:
  # If the enemy's HP is below 10 HP, you need to call self.enraged() which we created earlier.
  def isEnraged(self):
    # Your code here
    if self.HP < 10:
      self.enraged()
```

```python
# Now, with your knowledge of inheritence, you need to create two classes.

# Class Goblin: This class inherits Enemy. The parameters to be passed to the parent's constructor are:
# * name : 'Goblin'
# * maxHP : 50
# * attack : 15
# * defense : 5
# * maxMP : 0
# * carrying : A list containing a single Potion() item.

#----- Your code here -----#
class Goblin(Enemy):
  def __init__(self):
    super().__init__(Goblin,50,15,5,0, [Potion()])

# Class Orc: This class inherits Enemy. The parameters to be passed to the parent's constructor are:
# * name : 'Orc'
# * maxHP : 100
# * attack : 25
# * defense : 5
# * maxMP : 0
# * carrying : A list containing a Potion() item and a Elixir() item.
class Orc(Enemy):
  def __init__(self):
    super().__init__(Orc,100,25,5,0, [Potion(), Elixir()])
```

```python
# Necessary imports
import os
import pickle

# TASK 1
def checkIfCharacterSaved():
  # This function should check if there exists a file in the path 'player.pkl'.
  # If it exists, return True. Else, return False

  #--- Your code below here ---#
  if os.path.exists("player.pkl"):
    return True
  return False

# TASK 2

# Create a function 'saveCharacter', that takes a character object as parameter
# and saves it as a pickle object into the file 'player.pkl'.
# Hint 1: Open the filepath in write binary mode and dump() the pickle object into the file.
# Hint 2: The syntax for dump is pickle.dump(obj, file)

#--- Your code below here ---#
def saveCharacter(character):
  with open("player.pkl", 'wb') as fw:
    pickle.dump(character, fw)


# TASK 3
# Create a function 'loadCharacter', that takes no parameters.
# It should open the 'player.pkl' file in read binary mode and load the contents into a variable.
# Finally, return the variable.
def loadCharacter():
  player = None
  with open("player.pkl", 'rb') as fr:
    player = pickle.load(character, fr)
  return player
```

```python
#@title Just run this cell by clicking on the play button
def choose_character():
  print("Choose your character class:")
  print("1. Warrior")
  print("2. Mage")
  choice = input("Enter a number: ")
  if choice == "1":
    return Warrior(input("Enter your name: "))
  elif choice == "2":
    return Mage(input("Enter your name: "))
  else:
    print("Invalid choice.")
  return choose_character()

def use_item(character):
  print("These are the items in your inventory:")
  numItems = character.displayInventory()
  if numItems == 0:
    print("You do not have any items. Please select another option")
    return False
  selectedIndex = int(input("Select your item of choice based on the index: "))
  print("-------------------------------------------------------------------")
  if selectedIndex >= 0 and selectedIndex < numItems:
    character.consumeItem(selectedIndex)
  else:
    print("You gave an invalid index. Please choose again")
    use_item(character)

def enemy_turn(character, enemy):
  print("Now, it's your enemy's turn.")
  print("Enemy:", enemy)
  print(f"{enemy.name} attacked you.")
  attack = enemy.getAttack()
  lostHP = character.receiveDamage(attack)
  # print(f"You lost {lostHP} HP. ")

def choose_action(character, enemy):
  print("Choose an action:")
  print("1. Attack")
  print("2. Special attack")
  print("3. Use item")
  choice = input("Enter a number: ")
  print("-------------------------------------------------------------------")
  if choice == "1":
    damage = character.getAttack()
    lostHP = enemy.receiveDamage(damage)
    if enemy.isAlive():
      enemy_turn(character, enemy)
    return None
  elif choice == "2" and isinstance(character, Warrior):
    damage = character.slash()
    lostHP = enemy.receiveDamage(damage)
    if enemy.isAlive():
      enemy_turn(character, enemy)
    return None
  elif choice == "2" and isinstance(character, Mage):
    damage = character.fireball()
    lostHP = enemy.receiveDamage(damage)
    if enemy.isAlive():
      enemy_turn(character, enemy)
    return
  elif choice == "3":
    usedStatus = use_item(character)
    if enemy.isAlive() and usedStatus != False:
      enemy_turn(character, enemy)
    return
  else:
    print("Invalid choice.")
  return choose_action(character, enemy)

def battle(character, enemies):
  while len(enemies) != None:
    enemy = enemies[0]
    print("-------------------------------------------------------------------")
    print()
    print("You are fighting :", enemy)
    choose_action(character, enemy)
    if not enemy.isAlive():
      print(f"Congratulations! You defeated {enemy.name}.")
      character.addItem(enemy.carrying)
      print(f"""You looted the enemy's items.\nYour current inventory is {character.inventory}""")
      print("-------------------------------------------------------------------")
      enemies.remove(enemy)
    elif not character.isAlive():
      print("Oh no. You died... Better luck next time.")
      break
    print("Stats at the end of the turn.")
    print(f"Your status: {character}")
    print(enemy)
    response = input("Turn complete. Would you like to save and quit?(y/n): ")
    print("###################################################################")
    print()
    if response == "y":
      saveCharacter(character)
      break
```
