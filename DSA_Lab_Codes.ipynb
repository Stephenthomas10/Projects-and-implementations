```python
#ex 1- ARRAY IMP OF STACK
class Stack:
    def __init__(self, max_size):
        self.MAX = max_size
        self.stack = [None] * max_size
        self.top = -1

    def push(self, value):
        if self.isFull():
            print("Error: Overflow\n")
        else:
            self.top += 1
            self.stack[self.top] = value

    def isFull(self):
        return self.top == self.MAX - 1

    def pop(self):
        if self.isEmpty():
            print("Stack underflow")
            return None
        else:
            x = self.stack[self.top]
            self.top -= 1
            return x

    def isEmpty(self):
        return self.top == -1

    def display(self):
        for i in range(self.top, -1, -1):
            print(self.stack[i])

# Example usage:
max_size = 5  # Set the maximum size of the stack
stack = Stack(max_size)

stack.push(1)
stack.push(2)
stack.push(3)
stack.push(4)
stack.push(5)

stack.display()  # Output: 5 4 3 2 1

popped_value = stack.pop()
print("Popped value:", popped_value)  # Output: Popped value: 5

stack.display()  # Output: 4 3 2 1
```

```python
operators=set(['+','-','*','/','(',')','^'])
priority={'+':1,'-':1,'*':2,'/':2,'^':3}
def infixtopostfix(expression):
    stack=[]
    op=''
    for character in expression
        if character not in operators:
            op+=character
        elif character=='(':
            stack.append('(')
        elif character==')':
            while stack and stack[-1]!='(':
                op+=stack.pop()
            stack.pop()
        else:
            while stack and stack[-1]!='(' and priority[character]<=priority[stack[-1]]:
                op+=stack.pop()
            stack.append(character)
    while stack:
        op+=stack.pop()
    return op
expression=input("Enter the infix expression:")
print("infix notation",expression)
print("postfix notation",infixtopostfix(expression))
```

```python
class Queue:
    def __init__(self, max_size):
        self.MAX = max_size
        self.queue = [None] * max_size
        self.front = self.rear = -1

    def enqueue(self, value):
        if self.isFull():
            print("Error: Overflow\n")
        else:
            if self.front == -1:
                self.front = 0
            self.rear += 1
            self.queue[self.rear] = value

    def isFull(self):
        return self.rear == self.MAX - 1

    def dequeue(self):
        if self.isEmpty():
            print("Queue underflow")
            return None
        else:
            x = self.queue[self.front]
            if self.front == self.rear:
                self.front = self.rear = -1
            else:
                self.front += 1
            return x

    def isEmpty(self):
        return self.front == -1

    def display(self):
        for i in range(self.front, self.rear + 1):
            print(self.queue[i])

# Example usage of Queue:
max_size = 5  # Set the maximum size of the queue
queue = Queue(max_size)

queue.enqueue(1)
queue.enqueue(2)
queue.enqueue(3)
queue.enqueue(4)
queue.enqueue(5)

queue.display()  # Output: 1 2 3 4 5

dequeued_value = queue.dequeue()
print("Dequeued value:", dequeued_value)  # Output: Dequeued value: 1

queue.display()  # Output: 2 3 4 5

class CircularQueue(Queue):
    def enqueue(self, value):
        if self.isFull():
            print("Error: Overflow\n")
            return
        if self.rear == self.MAX - 1 and self.front == 0:
            self.rear = 0
        else:
            self.rear += 1
        self.queue[self.rear] = value

    def dequeue(self):
        if self.isEmpty():
            print("Queue underflow")
            return None
        x = self.queue[self.front]
        if self.front == self.rear:
            self.front = self.rear = -1
        elif self.front == self.MAX - 1:
            self.front = 0
        else:
            self.front += 1
        return x

# Example usage of CircularQueue:
max_size = 5  # Set the maximum size of the circular queue
circular_queue = CircularQueue(max_size)

circular_queue.enqueue(1)
circular_queue.enqueue(2)
circular_queue.enqueue(3)
circular_queue.enqueue(4)
circular_queue.enqueue(5)

circular_queue.display()  # Output: 1 2 3 4 5

dequeued_value = circular_queue.dequeue()
print("Dequeued value:", dequeued_value)  # Output: Dequeued value: 1

circular_queue.display()  # Output: 2 3 4 5
```

```python
class DoublyLinkedNode:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

class DoublyLinkedList:
    def __init__(self):
        self.firstNode = None
        self.lastNode = None

    def insertAfter(self, node, newNode):
        newNode.prev = node
        if node.next is None:
            newNode.next = None
            self.lastNode = newNode
        else:
            newNode.next = node.next
            node.next.prev = newNode
        node.next = newNode

    def insertBefore(self, node, newNode):
        newNode.next = node
        if node.prev is None:
            newNode.prev = None
            self.firstNode = newNode
        else:
            newNode.prev = node.prev
            node.prev.next = newNode
        node.prev = newNode

    def insertBeginning(self, newNode):
        if self.firstNode is None:
            self.firstNode = newNode
            self.lastNode = newNode
            newNode.prev = None
            newNode.next = None
        else:
            self.insertBefore(self.firstNode, newNode)

    def insertEnd(self, newNode):
        if self.lastNode is None:
            self.insertBeginning(newNode)
        else:
            self.insertAfter(self.lastNode, newNode)

    def display(self):
        temp = self.firstNode
        while temp:
            print(temp.data, end=" <-> ")
            temp = temp.next
        print("None")

# Example usage of Doubly Linked List:
dll = DoublyLinkedList()

node1 = DoublyLinkedNode(1)
node2 = DoublyLinkedNode(2)
node3 = DoublyLinkedNode(3)

dll.insertBeginning(node1)
dll.insertEnd(node3)
dll.insertAfter(node1, node2)

dll.display()  # Output: 1 <-> 2 <-> 3 <-> None
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.link = None

class LinkedStack:
    def __init__(self):
        self.top = None

    def push(self, item):
        temp = Node(item)
        temp.link = self.top
        self.top = temp

    def pop(self):
        if self.top is None:
            print("Stack is empty")
            return None
        temp = self.top
        item = temp.data
        self.top = self.top.link
        del temp
        return item

# Example usage of Linked Stack:
stack = LinkedStack()

stack.push(1)
stack.push(2)
stack.push(3)

print(stack.pop())  # Output: 3
print(stack.pop())  # Output: 2
print(stack.pop())  # Output: 1
```

```python
class LinkedQueue:
    def __init__(self):
        self.front = None
        self.rear = None

    def enqueue(self, item):
        temp = Node(item)
        temp.link = None
        if self.front is None:
            self.rear = self.front = temp
            return
        self.rear.link = temp
        self.rear = self.rear.link

    def dequeue(self):
        if self.front is None:
            print("Queue is empty")
            return None
        item = self.front.data
        temp = self.front
        self.front = self.front.link
        del temp
        return item

# Example usage of Linked Queue:
queue = LinkedQueue()

queue.enqueue(1)
queue.enqueue(2)
queue.enqueue(3)

print(queue.dequeue())  # Output: 1
print(queue.dequeue())  # Output: 2
print(queue.dequeue())  # Output: 3
```

```python
def selection_sort(arr):
    n = len(arr)
    for j in range(n - 1):
        smallest = j
        for i in range(j + 1, n):
            if arr[i] < arr[smallest]:
                smallest = i
        arr[j], arr[smallest] = arr[smallest], arr[j]

# Example usage of Selection Sort:
arr = [64, 25, 12, 22, 11]
selection_sort(arr)
print("Sorted array:", arr)
```

```python
def linear_search(arr, key):
    n = len(arr)
    for i in range(n):
        if arr[i] == key:
            return i
    return -1

# Example usage of Linear Search:
arr = [64, 25, 12, 22, 11]
key = 22
result = linear_search(arr, key)
if result != -1:
    print(f"Element {key} found at index {result}")
else:
    print(f"Element {key} not found")
```

```python
def binary_search(arr, key):
    left = 0
    right = len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == key:
            return mid
        elif key < arr[mid]:
            right = mid - 1
        else:
            left = mid + 1
    return -1

# Example usage of Binary Search:
arr = [11, 12, 22, 25, 64]
key = 22
result = binary_search(arr, key)
if result != -1:
    print(f"Element {key} found at index {result}")
else:
    print(f"Element {key} not found")
```

```python
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        B = arr[:mid]
        C = arr[mid:]

        merge_sort(B)
        merge_sort(C)

        i = j = k = 0

        while i < len(B) and j < len(C):
            if B[i] <= C[j]:
                arr[k] = B[i]
                i += 1
            else:
                arr[k] = C[j]
                j += 1
            k += 1

        while i < len(B):
            arr[k] = B[i]
            i += 1
            k += 1

        while j < len(C):
            arr[k] = C[j]
            j += 1
            k += 1

# Example usage of Merge Sort:
arr = [64, 25, 12, 22, 11]
merge_sort(arr)
print("Sorted array:", arr)
```

```python
def quick_sort(arr):
    def partition(arr, l, r):
        pivot = arr[l]
        i, j = l + 1, r
        while True:
            while i <= j and arr[i] <= pivot:
                i += 1
            while i <= j and arr[j] >= pivot:
                j -= 1
            if i <= j:
                arr[i], arr[j] = arr[j], arr[i]
            else:
                break
        arr[l], arr[j] = arr[j], arr[l]
        return j

    def quicksort(arr, l, r):
        if l < r:
            s = partition(arr, l, r)
            quicksort(arr, l, s - 1)
            quicksort(arr, s + 1, r)

    quicksort(arr, 0, len(arr) - 1)

# Example usage of Quick Sort:
arr = [64, 25, 12, 22, 11]
quick_sort(arr)
print("Sorted array:", arr)
```

```python
class TreeNode:
    def __init__(self, data):
        self.data = data
        self.lchild = None
        self.rchild = None

def insert_bst(root, new_node):
    if new_node.data < root.data:
        if root.lchild is None:
            root.lchild = new_node
        else:
            insert_bst(root.lchild, new_node)
    if new_node.data > root.data:
        if root.rchild is None:
            root.rchild = new_node
        else:
            insert_bst(root.rchild, new_node)

def in_order_bst(root):
    if root is not None:
        in_order_bst(root.lchild)
        print(root.data, end=" ")
        in_order_bst(root.rchild)

def pre_order_bst(root):
    if root is not None:
        print(root.data, end=" ")
        pre_order_bst(root.lchild)
        pre_order_bst(root.rchild)

def post_order_bst(root):
    if root is not None:
        post_order_bst(root.lchild)
        post_order_bst(root.rchild)
        print(root.data, end=" ")

from collections import deque

def level_order_bst(root):
    if root is None:
        return

    queue = deque()
    queue.append(root)

    while queue:
        node = queue.popleft()
        if node:
            print(node.data, end=" ")
            queue.append(node.lchild)
            queue.append(node.rchild)

# Example usage of Binary Search Tree and Traversals:
root = TreeNode(50)
insert_bst(root, TreeNode(30))
insert_bst(root, TreeNode(70))
insert_bst(root, TreeNode(20))
insert_bst(root, TreeNode(40))
insert_bst(root, TreeNode(60))
insert_bst(root, TreeNode(80))

print("In-order traversal:")
in_order_bst(root)
print("\nPre-order traversal:")
pre_order_bst(root)
print("\nPost-order traversal:")
post_order_bst(root)
print("\nLevel-order traversal:")
level_order_bst(root)
```
